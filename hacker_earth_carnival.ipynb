```python
y=train1['Selling_Price']
x=train1.drop(['Selling_Price'],axis=1)
```

```python
from xgboost import  XGBRFRegressor
```

```python
clf = XGBRFRegressor()
```

```python
clf.fit(x_train,y_train)
```

```python
y_pred_train_xgb = clf.predict(x_train)
y_pred_valid_xgb = clf.predict(x_val)
```

```python
k=pd.DataFrame(y_pred_train_xgb,columns=['pred_train_xgb'])
```

```python
k.pred_train_xgb.sort_values()
```

```python
j=pd.DataFrame(y_pred_valid_xgb,columns=['pred_valid_xgb'])
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
print(np.sqrt(mean_squared_error(y_val,y_pred_valid_xgb)))
print(np.sqrt(mean_squared_error(y_train,y_pred_train_xgb)))
```
